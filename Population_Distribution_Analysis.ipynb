```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create sample population dataset
data = {
    'Age_Group': ['0-18', '19-30', '31-45', '46-60', '60+'],
    'Population': [500, 700, 650, 400, 250]
}

# Create DataFrame
df = pd.DataFrame(data)

# Show dataset
print("Population Dataset:")
print(df)

# -----------------------------
# Bar Chart
# -----------------------------
plt.figure(figsize=(8,5))

sns.barplot(
    x='Age_Group',
    y='Population',
    data=df
)

plt.title("Population Distribution by Age Group")

plt.xlabel("Age Group")
plt.ylabel("Population")

plt.show()

# -----------------------------
# Histogram
# -----------------------------
plt.figure(figsize=(8,5))

plt.hist(df['Population'], bins=5)

plt.title("Histogram of Population Distribution")

plt.xlabel("Population")
plt.ylabel("Frequency")

plt.show()
```
